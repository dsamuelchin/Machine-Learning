```python
df['bmi_class'] = pd.cut(df['bmi'], bins=[0, 18.5, 25, 30,100],
                         labels=['Underweight', 'Normal', 'Overweight', 'Obese'])
```

```python
mappings = {
    'gender': {0: 'Male', 1: 'Female'},
    'cvd': {0: 'No', 1: 'Yes'},
    'afb': {0: 'No', 1: 'Yes'},
    'sho': {0: 'No', 1: 'Yes'},
    'chf': {0: 'No', 1: 'Yes'},
    'av3': {0: 'No', 1: 'Yes'},
    'miord': {0: 'First', 1: 'Recurrent'},
    'mitype': {0: 'non Q-wave', 1: 'Q-wave'},
    'year': {1: '1997', 2: '1999', 3: '2001'},
    'dstat': {0: 'Alive', 1: 'Deceased'},
    'fstat': {0: 'Alive', 1: 'Deceased'}
}

for col, mapping in mappings.items():
    df[f'{col}_label'] = df[col].map(mapping)

df[[col + '_label' for col in mappings.keys()]].head()
```

Creating Visuals

```python

numeric_cols = ['age', 'hr', 'sysbp', 'diasbp', 'bmi', 'los', 'lenfol']

fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    sns.histplot(df[col], kde=True, ax=axes[i], color='teal')
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel('')

for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.savefig('numeric_distributions.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
numeric_cols_for_boxplot = ['age', 'hr', 'sysbp', 'diasbp', 'bmi', 'los', 'lenfol']
n_rows = 3
n_cols = 3

fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(n_cols * 6, n_rows * 5))
axes = axes.flatten()

fstat_palette = {'Alive': 'forestgreen', 'Deceased': 'firebrick'}

for i, col in enumerate(numeric_cols_for_boxplot):
    sns.boxplot(data=df, x='fstat_label', y=col, ax=axes[i], hue='fstat_label', palette=fstat_palette, order=['Alive', 'Deceased'], legend=False)
    axes[i].set_title(f'Distribution of {col} by Fstat')
    axes[i].set_xlabel('Vital Status at Last Follow-up')
    axes[i].set_ylabel(col)

for j in range(len(numeric_cols_for_boxplot), len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.suptitle('Boxplots of Numeric Variables by Vital Status at Last Follow-up', y=1.02, fontsize=16)
plt.savefig('vital_status_boxplots.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import math
categorical_cols_for_fstat_comparison = [
    'gender_label', 'cvd_label', 'afb_label', 'sho_label', 'chf_label',
    'av3_label', 'miord_label', 'mitype_label', 'year_label', 'bmi_class', 'dstat_label'
]

num_plots = len(categorical_cols_for_fstat_comparison)
n_cols = 3
n_rows = math.ceil(num_plots / n_cols)
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(n_cols * 5, n_rows * 5))
axes = axes.flatten()

fstat_palette = {'Alive': 'forestgreen', 'Deceased': 'firebrick'}

for i, col in enumerate(categorical_cols_for_fstat_comparison):
    if col in df.columns:
        sns.countplot(data=df, x=col, hue='fstat_label', ax=axes[i],
                      palette=fstat_palette, hue_order=['Alive', 'Deceased'])

        plot_title = col.replace('_label', '').replace('_', ' ').title()
        if col == 'dstat_label': plot_title = 'Vital Status at Discharge'

        axes[i].set_title(f'{plot_title} vs. Fstat', fontsize=12)
        axes[i].set_xlabel('')
        axes[i].tick_params(axis='x', rotation=45)
    else:
        axes[i].set_title(f"Column {col} missing")

for j in range(num_plots, len(axes)):
    axes[j].axis('off')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.suptitle('Vital Status (Fstat) Distribution Across Categorical Variables', fontsize=16)
plt.savefig('Vital Status BarGraph .png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler

features = ['age', 'hr', 'bmi', 'sho','chf', 'diasbp']
X = df[features].copy()

scaler = StandardScaler()
X[['age', 'hr', 'bmi', 'diasbp']] = scaler.fit_transform(X[['age', 'hr', 'bmi', 'diasbp']])

X = sm.add_constant(X)

y = df['fstat']
model = sm.Logit(y, X).fit()
print(model.summary())
```

```python
from sklearn.metrics import confusion_matrix, classification_report

probabilities = model.predict(X)

predictions = (probabilities > 0.5).astype(int)

cm = confusion_matrix(y, predictions)
print("\n--- Confusion Matrix ---")
print(cm)

print("\n--- Classification Report ---")
print(classification_report(y, predictions))
```
